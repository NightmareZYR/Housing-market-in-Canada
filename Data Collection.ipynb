The informaton can be collected are "ic-beds", "ic-baths", "ic-sqft", "ic-lotsize", "property-type ic-proptype", "price" and "item-address"

Let's store them in the data set

```python
data = {}
soups = {}
tables = []
# Create a df to store info
Point2_data = pd.DataFrame(columns=['Beds','Bath','Sqft','Lot_Size','Type','Price','Address'])
for i in range(1,23):
    url='https://www.point2homes.com/CA/Real-Estate-Maps.html?location=Nova+Scotia&PropertyType=House&search_mode=location&page=' + str(i) + '&SelectedView=map_view&LocationGeoId=205403&viewport=-67.2614835734375%2C42.61290244283285%2C-58.9228605265625%2C47.12543080145628&zoom=7&location_changed=&ajax=1'
    data = requests.get(url).text
    soup = BeautifulSoup(data, 'html.parser')
    tables.append(soup.find_all("div", class_="item-details"))
    
    # extract information from each table
    for table in (tables[(i-1)]):
        if (len(str(table.find_all("li", class_='ic-beds')).split())<3):
            Nbed = "NaN"
        else:
            Nbed = str(table.find_all("li", class_='ic-beds')).split()[3][8:-9]
        
        if (len(str(table.find_all("li", class_='ic-baths')).split())<3):
            NBath = "NaN"
        else:
            NBath = str(table.find_all("li", class_='ic-baths')).split()[3][8:-9]
        
        if (len(str(table.find_all("li", class_='ic-sqft')).split())<3):
            NSqft = "NaN"
        else:
            NSqft = str(table.find_all("li", class_='ic-sqft')).split()[3][8:-9]
    
        if (len(str(table.find_all("li", class_='ic-lotsize')).split())<4):
            NLot_Size = "NaN"
        else:
            NLot_Size = str(table.find_all("li", class_='ic-lotsize')).split()[4][-4:]
        
        if (len(str(table.find_all("li", class_='ic-proptype')).split())<3):
            NType = "NaN"
        else:
            NType = str(table.find_all("li", class_='ic-proptype')).split()[3]


        NPrice = table.find("div", class_='price')['data-price']
        NAddress = table.find("div", class_='item-address')['data-address']
    # append data into the table 
        Point2_data = Point2_data.append({'Beds':Nbed,'Bath':NBath,'Sqft':NSqft,'Lot_Size':NLot_Size,'Type':NType,'Price':NPrice,'Address':NAddress},ignore_index=True)
```

Now I have the data, so we need to clean it.

Column "Sqft", 'Lot_Size', and 'Price' need to be reformated

```python
for i in range(2121):
    Point2_data['Sqft'][i] = Point2_data['Sqft'][i].replace(',','')
```

```python
for i in range(2121):
    Point2_data['Lot_Size'][i] = Point2_data['Lot_Size'][i].replace(',','')
    Point2_data['Lot_Size'][i] = Point2_data['Lot_Size'][i].replace('>','')
```

```python
for i in range(2121):
    Point2_data['Price'][i] = Point2_data['Price'][i].replace(',','')
    Point2_data['Price'][i] = Point2_data['Price'][i].replace('$','')
    Point2_data['Price'][i] = Point2_data['Price'][i].replace('CAD','')
```

```python
Point2_data
```

All the information is stored as string, so for some of columns, I need to convert to float.

Also, I need to find index of information need to be edited.

```python
edited_index={}
listBed=[]
listBath=[]
listSqft=[]
listLot_Size=[]
listPrice=[]
for i in range(2121):
    # Beds column
    try:
        Point2_data["Beds"][i]=float(Point2_data["Beds"][i])
    except(ValueError):
        listBed.append(i)
        edited_index.update({'Bed': listBed})
    pass
    # Bath column
    try:
        Point2_data['Bath'][i]=float(Point2_data['Bath'][i])
    except(ValueError):
        listBath.append(i)
        edited_index.update({'Bath': listBath})
    pass
    # Sqft column
    try:
        Point2_data['Sqft'][i]=float(Point2_data['Sqft'][i])
    except(ValueError):
        listSqft.append(i)
        edited_index.update({'Sqft': listSqft})
    pass
    # Lot_Size column
    try:
        Point2_data['Lot_Size'][i]=float(Point2_data['Lot_Size'][i])
    except(ValueError):
        listLot_Size.append(i)
        edited_index.update({'Lot_Size': listLot_Size})
    pass
    # Price column
    try:
        Point2_data['Price'][i]=float(Point2_data['Price'][i])
    except(ValueError):
        listLot_Size.append(i)
        edited_index.update({'Price': listLot_Size})
    pass
```
