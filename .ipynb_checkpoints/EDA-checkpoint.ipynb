### Now I need to standardlize the data.

```python
from sklearn.preprocessing import StandardScaler
X = cluster_Point2_data.values[:,:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

### Modeling

```python
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
cluster_Point2_data["Clus_km"] = labels
cluster_Point2_data.head(100)
```

```python
# create map and display it
Canada_map = folium.Map(location=[56.1304, -92.3468], zoom_start=4)        
labels=[]
bedrooms = pdf['Beds']
bathrooms = pdf['Bath']
sqft = pdf['Sqft']
price = pdf['Price']
for i in range(len(pdf['Beds'])):
    labels.append (str(bedrooms[i]) + ' bedrooms' + ","
                 + str(bathrooms[i]) + ' bathrooms' + ","
                 + str(sqft[i]) + ' sqfts' + ","
                 + '$' + str(price[i]) )          
# loop through the 100 crimes and add each to the map
for lat, lng, label in zip(pdf['latitude'], pdf['longitude'], labels):
    folium.features.CircleMarker(
        [lat, lng],
        radius=2, # define how big you want the circle markers to be
        color='red',
        fill=False,
        popup=label,
        fill_color='blue',
        fill_opacity=0.6
    ).add_to(Canada_map)
Canada_map
```
