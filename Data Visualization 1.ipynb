```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print('Matplotlib version: ', mpl.__version__) # >= 2.0.0
import numpy as np
```

Let's look at Bedrooms and Bathrooms

```python
Beds = Point2_data['Beds'].value_counts()
Baths = Point2_data['Bath'].value_counts()
df = pd.DataFrame({'Beds': Beds,'Baths': Baths})
```

```python
Beds.plot(kind='barh',figsize=(16,6))
plt.title('Canada housing market with Bedrooms')
plt.xlabel('Number of houses')
plt.ylabel('Number of bedrooms')
for index, value in enumerate(Beds):
    plt.text(value, index, str(value))
```

#### Three bedrooms and four bedrooms are the most common type of houses.
    
Let's look at bathroom

```python
ax = df['Baths'].plot.barh(figsize=(16,6), color='C1')
ax.set_xlabel('Number of houses')
ax.set_ylabel('Number of bathrooms')
ax.set_title('Houses with Bathroom')
for i, v in enumerate(df['Baths']):
    ax.text(v +2, i -0.1 , str(v), color = 'C3', fontweight = 'bold')
```

#### Two bedrooms are the most common type of house, then three bedrooms are popular as well.

Let's see them together

```python
fig = plt.figure() # create figure

ax0 = fig.add_subplot(1, 2, 1) # add subplot 1 (1 row, 2 columns, first plot)
ax1 = fig.add_subplot(1, 2, 2) # add subplot 2 (1 row, 2 columns, second plot). See tip below**

# Subplot 1: 
df['Beds'].plot(kind='barh', color='C1', figsize=(20, 5), ax=ax0) # add to subplot 1
ax0.set_title('Houses in Nova Socita with Bedrooms')
ax0.set_ylabel('Number of bedrooms')
ax0.set_xlabel('Number of houses')
for i, v in enumerate(df['Beds']):
    ax0.text(v +2, i -0.1 , str(v), color = 'C3', fontweight = 'bold')
    
# Subplot 2: 
df['Baths'].plot(kind='barh',  color='C0', figsize=(20, 5), ax=ax1) # add to subplot 2
ax1.set_title ('Houses in Nova Socita with Bathrooms')
ax1.set_ylabel('Number of bathrooms')
ax1.set_xlabel('Number of houses')
for i, v in enumerate(df['Baths']):
    ax1.text(v +2, i -0.1 , str(v), color = 'C3', fontweight = 'bold')
plt.show()
```

### Now I want to see what combination of bedrooms and bathrooms are popular

```python
df_bb = Point2_data[['Beds','Bath']]
```

```python
df_bb = df_bb.groupby(['Beds','Bath']).size()
```

```python
df_buble = pd.DataFrame({'Counts':df_bb})
```

```python
df_buble = df_buble.reset_index()
```

```python
norm = (df_buble['Counts'] - df_buble['Counts'].min()) / (df_buble['Counts'].max() - df_buble['Counts'].min())
```

```python
ax = df_buble.plot(kind='scatter',
                    x='Beds',
                    y='Bath',
                    figsize=(18, 8),
                    alpha=0.6,  # transparency
                    c = 'C1',
                    s=norm * 2000 + 10,  # pass in weights 
                    xlim=(0, 17),
                    ylim=(0,20)
             )
ax.set_yticks(np.arange(0, 20, step=1))
ax.set_xticks(np.arange(0, 17, step=1))
ax.set_title('Counts of bedroom & bathroom combo')
ax.set_xlabel('Number of bedrooms')
ax.set_ylabel('Number of bathrooms')
```

It is not hard to find that **three bedrooms with two Bathroom**  is the most type of houses in the market.

It is likely that **adding one more bedroom is *cheaper* than adding one more bathroom**

### Now I want to see how much is gonna cost to buy a sqft in different bedroom-bathroom combo

```python
df_pps = Point2_data[['Beds','Bath','Price per sqft']]
```

```python
df_pps = df_pps.groupby(['Beds','Bath']).mean().reset_index()
```

```python
df_pps.head()
```

```python
df_pps['Price per sqft']= df_pps['Price per sqft'].astype(int)
```

```python
norm3 = (df_pps['Price per sqft'] - df_pps['Price per sqft'].min()) / (df_pps['Price per sqft'].max() - df_pps['Price per sqft'].min())
```

```python
ax2 = df_pps.plot(kind='scatter',
                    x='Beds',
                    y='Bath',
                    figsize=(18, 8),
                    alpha=0.6,  # transparency
                    c = 'C1',
                    s=norm3 * 2000,  # pass in weights 
                    xlim=(0, 17),
                    ylim=(0,20)
             )
ax2.set_yticks(np.arange(0, 20, step=1))
ax2.set_xticks(np.arange(0, 17, step=1))
ax2.set_title('Average price per sqft of bedroom & bathroom combo')
ax2.set_xlabel('Number of bedrooms')
ax2.set_ylabel('Number of bathrooms')
```

It likely that price per sqft of one bedroom is much expensive than other combo with more than one bedroom.

#### Check out 'the price per sqft', 'the price per bedroom',and 'the price per bathroom' in different province

```python
df_pps_pro = Point2_data[['Province','Price per sqft']]
```

```python
df_pps_pro = df_pps_pro.groupby(['Province']).mean().reset_index()
```

```python
df_pps_pro['Price per sqft'] = df_pps_pro['Price per sqft'].astype(int)
```

```python
df_pps_pro = df_pps_pro.sort_values(['Price per sqft'],ascending=False).reset_index(drop=True)
```

```python
import seaborn as sns
```

```python
fig_dims = (13, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ax3 = sns.barplot(data=df_pps_pro, x='Province',y= 'Price per sqft')
for i, v in enumerate(df_pps_pro["Price per sqft"].iteritems()):        
    ax3.text(i ,v[1], "{:,}".format(v[1]), color='b', va ='bottom')
ax3.set_title("Price per square feet in Canada")
```

```python
Point2_data['price_per_bed'] = Point2_data['Price']/Point2_data['Beds']
```

```python
df_ppb_pro = Point2_data[['Province','price_per_bed']]
```

```python
df_ppb_pro = df_ppb_pro.groupby(['Province']).mean().reset_index()
```

```python
df_ppb_pro['price_per_bed'] = df_ppb_pro['price_per_bed'].astype(int)
df_ppb_pro = df_ppb_pro.sort_values(['price_per_bed'],ascending=False).reset_index(drop=True)
```

```python
fig_dims = (13, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ax4 = sns.barplot(data=df_ppb_pro, x='Province',y= 'price_per_bed')
for i, v in enumerate(df_ppb_pro["price_per_bed"].iteritems()):        
    ax4.text(i ,v[1], "{:,}".format(v[1]), color='b', va ='bottom')
ax4.set_title("Price per bedroom in Canada")
```

```python
Point2_data['price_per_bath'] = Point2_data['Price']/Point2_data['Bath']
```

```python
df_ppbath_pro = Point2_data[['Province','price_per_bath']]
```

```python
df_ppbath_pro = df_ppbath_pro.groupby(['Province']).mean().reset_index()
```

```python
df_ppbath_pro['price_per_bath'] = df_ppbath_pro['price_per_bath'].astype(int)
df_ppbath_pro = df_ppbath_pro.sort_values(['price_per_bath'],ascending=False).reset_index(drop=True)
```

```python
fig_dims = (13, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ax4 = sns.barplot(data=df_ppbath_pro, x='Province',y= 'price_per_bath')
for i, v in enumerate(df_ppbath_pro["price_per_bath"].iteritems()):        
    ax4.text(i ,v[1], "{:,}".format(v[1]), color='b', va ='bottom')
ax4.set_title("Price per bathroom in Canada")
```
